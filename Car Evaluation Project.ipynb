```python
x = dataSet.drop(['class'], axis = 1)
y = dataSet['class']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.65, test_size = 0.25, random_state = 210)
print("X_train: {}".format(x_train.shape))
print("X_test: {}".format(x_test.shape))
print("Y_train: {}".format(y_train.shape))
print("Y_test: {}".format(y_test.shape))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth = 8) 
dt.fit(x_train,y_train)

yp_train = dt.predict(x_train)
yp_test = dt.predict(x_test)

from sklearn.metrics import accuracy_score, classification_report
def experiment_results(y_train,yp_train,y_test,yp_test):
  print("--------------------------------------------------------------------------")
  print("Đánh giá kết quả phân lớp trên bộ dữ liệu huấn luyện")
  print(classification_report(y_train, yp_train))
  print("Đánh giá kết quả phân lớp trên bộ dữ liệu thực nghiệm")
  print(classification_report(y_test, yp_test))
  print("--------------------------------------------------------------------------")
  # Accuracy
  print("Độ chính xác của dự đoán trên dữ liệu huấn luyện: {}".format(round(accuracy_score(y_train,yp_train),2)))
  print("Độ chính xác của dự đoán trên dữ liệu thực nghiệm: {}".format(round(accuracy_score(y_test,yp_test),2)))
  print("--------------------------------------------------------------------------")

experiment_results(y_train,yp_train,y_test,yp_test)
```
